```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_range_list = []

for n in range(89):
    x_range_val = 0.06 + n*0.005
    x_range_list.append(x_range_val)
    
N = 6.02E22
a = 4*10**(-49)
b = 6*10**(-29)
KT = 0.267 * (a/b)

Gibb_list = []
for n in range(89):
    Gibb_1 = -N*KT*np.log(x_range_list[n] - N*b) 
    Gibb_2 = (N*KT*N*b)/(x_range_list[n] - N*b) 
    Gibb_3 = (2*a*N**2)/(x_range_list[n])
    Gibb = Gibb_1 + Gibb_2 - Gibb_3
   
    Gibb_list.append(Gibb)
    

p_list = []
for n in range(89):
    p_1 = N*KT/(-N*b+x_range_list[n]) 
    p_2 = (a*N**2)/x_range_list[n]
    p = p_1 - p_2
    p_list.append(p)

p_c = a/(27*b**2)


print(p_c)
p_over_list = []
for n in range(89):
    p_over = p_list[n]/4115226.3374485606
    p_over_list.append(p_over)

plt.figure(1)
plt.plot(p_over_list,Gibb_list)
plt.grid(True)
plt.title("Gibb vs p")
plt.xlabel("p/p_o")
plt.ylabel("Gibb")
print(Gibb_list)

V_c = 3*b

v_over_list = []
for n in range(89):
    v_over = x_range_list[n]/(6e-26)
    v_over_list.append(v_over)
    
plt.figure(2)
plt.plot(p_over_list,v_over_list)
plt.title("v vs p")
plt.xlabel("p/p_o")
plt.ylabel("v/v_o")
plt.grid(True)
```
